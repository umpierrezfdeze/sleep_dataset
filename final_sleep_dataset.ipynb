**Exploración de variables categóricas**

Analizamos los valores únicos de cada variable categórica para detectar inconsistencias, categorías duplicadas o valores que necesiten unificarse.

```python
# Valores únicos por columna categórica
for col in ['Gender', 'Occupation', 'BMI Category', 'Sleep Disorder']:
    print(f'{col}')
    print(df[col].value_counts())
    print()
```

La exploración de variables categóricas revela tres problemas:

1. Occupation: 'Salesperson' (32) y 'Sales Representative' (2)
   son la misma categoría con nombres distintos.
   Las unificaremos en 'Sales Representative'.

2. BMI Category: 'Normal' (195) y 'Normal Weight' (21)
   representan lo mismo con nombres distintos.
   Las unificaremos en 'Normal'.

3. Sleep Disorder: los 219 NaN detectados antes no aparecen
   aquí porque value_counts() los ignora por defecto.
   Corresponden a personas sin trastorno del sueño y los
   reemplazaremos por 'None'.

Estos tres problemas se corregirán en la fase de limpieza.

**Limpieza de datos**

Corregimos los problemas detectados en la exploración. Trabajamos sobre una copia del dataframe original para no modificar los datos en bruto.

```python
df_clean = df.copy()

print('Copia creada')
```

**Sleep Disorder** — reemplazar NaN por 'None'

Los 219 valores NaN en Sleep Disorder no son datos perdidos, sino personas que no tienen ningún trastorno del sueño. Los reemplazamos por 'None' para que sean analizables.

```python
df_clean['Sleep Disorder'] = df_clean['Sleep Disorder'].fillna('None')

print('Sleep Disorder después de la limpieza:')
print(df_clean['Sleep Disorder'].value_counts())
```

**BMI Category — unificar 'Normal' y 'Normal Weight'**

'Normal' y 'Normal Weight' representan la misma categoría
con nombres distintos. Las unificamos en 'Normal'.

```python
df_clean['BMI Category'] = df_clean['BMI Category'].replace('Normal Weight', 'Normal')

print('BMI Category después de la limpieza:')
print(df_clean['BMI Category'].value_counts())
```

**Occupation — unificar 'Salesperson' y 'Sales Representative'**

'Salesperson' y 'Sales Representative' son la misma ocupación
con nombres distintos. Las unificamos en 'Sales Representative'.

```python
df_clean['Occupation'] = df_clean['Occupation'].replace('Salesperson', 'Sales Representative')

print('Occupation después de la limpieza:')
print(df_clean['Occupation'].value_counts())
```

**Blood Pressure — separar en dos columnas numéricas**

La columna Blood Pressure tiene formato string '120/80'
(sistólica/diastólica) y no es analizable directamente.

La separamos en dos columnas numéricas: bp_systolic y bp_diastolic.

```python
df_clean[['bp_systolic', 'bp_diastolic']] = (
    df_clean['Blood Pressure']
    .str.split('/', expand=True)
    .astype(int)
)

df_clean = df_clean.drop(columns=['Blood Pressure'])

print('Nuevas columnas de presión arterial:')
print(df_clean[['bp_systolic', 'bp_diastolic']].describe().round(1))
```

**Renombrar columnas a snake_case**

Renombramos las columnas eliminando espacios y pasándolas
a minúsculas para seguir las convenciones de Python
y facilitar el trabajo posterior con el dataset.

```python
df_clean.columns = [
    'person_id', 'gender', 'age', 'occupation', 'sleep_duration',
    'sleep_quality', 'physical_activity', 'stress_level',
    'bmi_category', 'heart_rate', 'daily_steps',
    'sleep_disorder', 'bp_systolic', 'bp_diastolic'
]

print('Columnas renombradas:')
print(list(df_clean.columns))
```

**Validación final**

Comprobamos que todos los pasos de limpieza se han aplicado
correctamente y el dataset está listo para el análisis.

```python
# Verificación de nulos
print('=== Valores nulos ===')
print(df_clean.isnull().sum())

# Verificación de duplicados
print(f'\n=== Duplicados ===')
print(f'Filas duplicadas: {df_clean.duplicated().sum()}')

# Resumen final
print(f'\n=== Resumen final ===')
print(f'Dimensiones: {df_clean.shape[0]} filas x {df_clean.shape[1]} columnas')
print(f'\nColumnas: {list(df_clean.columns)}')
print(f'\nTipos de datos:')
print(df_clean.dtypes)
```

**Resultado de la validación**

- 0 valores nulos en todas las columnas
- 242 filas comparten valores en sus variables pero NO son
  duplicados reales, ya que cada fila tiene un Person ID
  único que identifica a una persona distinta
- 374 filas x 13 columnas
- Todos los tipos de datos son correctos:
  numéricas como int64/float64 y categóricas como object


El dataset está limpio y listo para el análisis.
